```python
print(train.info())
train.index = train['Date']
del train['Date']
test.index = test['Date']
del test['Date']
print(train.info())
```

```python
print('Dimensions : \ntrain set: {}\ntest set : {}'.format(train.shape,test.shape))
```

```python
train.sample(5)
```

```python
test.head(5)
```

```python
trainX = train.drop(['Close'], axis=1)
trainy = train['Close']
testX = test.drop(['Close'], axis=1)
testy = test['Close']
print('Dimensions : \ntrain set: {}\ntrainX set : {}\ntrainy set : {}\ntest set : {}\ntestX set : {}\ntesty set : {}'
      .format(train.shape,trainX.shape,trainy.shape,test.shape,testX.shape,testy.shape))
```

```python
# Scaling the values for faster training of the models.
from sklearn.preprocessing import MinMaxScaler

scaledtrain = trainX
scalers={}

for i in trainX.columns:
    scaler = MinMaxScaler(feature_range=(0,1))
    s_s = scaler.fit_transform(scaledtrain[i].values.reshape(-1,1))
    s_s=np.reshape(s_s,len(s_s))
    scalers['scaler_'+ i] = scaler
    scaledtrain[i]=s_s

scaledtest = testX
for i in testX.columns:
    scaler = scalers['scaler_'+i]
    s_s = scaler.transform(scaledtest[i].values.reshape(-1,1))
    s_s=np.reshape(s_s,len(s_s))
    scalers['scaler_'+i] = scaler
    scaledtest[i]=s_s
```

```python
scaledtrain.head(5)
```

```python
trainX = scaledtrain
testX = scaledtest
```

# STAGE 4: Prediction

### Prediction Model 1 : Linear Regression

```python
from sklearn.linear_model import LinearRegression
modelLinearReg = LinearRegression()
modelLinearReg.fit(trainX,trainy)
predy = modelLinearReg.predict(testX)
```

```python
from math import sqrt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(testy,predy)
print('MAE:', mae)

rmse = sqrt(mean_squared_error(testy,predy))
print('RMSE:', rmse)

score = r2_score(testy, predy)
print("Accuracy: %.2f%%" % (score * 100.0))
```

```python
testX['Predictions'] = 0
testX['Predictions'] = predy

plt.figure(figsize = (16,10))
sns.lineplot(data=test[test['IndexName_Nifty 50'] == 1], x=test.index, y='Close', label = 'Actual Close price')
plt.plot(testX['Predictions'], label = 'Predictions')
plt.legend()
```

```python
print('Dimensions : \ntrainX set: {}\ntestX set : {}'.format(trainX.shape,testX.shape))
testX.drop(['Predictions'], axis=1,inplace=True)
print('Dimensions : \ntrainX set: {}\ntestX set : {}'.format(trainX.shape,testX.shape))
```
